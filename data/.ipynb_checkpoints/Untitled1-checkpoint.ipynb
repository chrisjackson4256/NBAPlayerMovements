```python
import os
import py7zlib
```

```python
tree = "/Users/c.b.jackson/Google Drive/NBA-Player-Movements/data/2016.NBA.Raw.SportVU.Game.Logs/"
dst = "/Users/c.b.jackson/Google Drive/NBA-Player-Movements/D3/data/"

for dirpath, dirname, filename in os.walk(tree):
    for myfile in filename:
        if myfile.endswith('.7z') and (not myfile.startswith('2016')):
            myfile_name = myfile[:-2] + "json"
            print(myfile_name)
            myZip = py7zlib.Archive7z(open(os.path.join(dirpath,myfile), 'rb'))
            csvInZipFile = zip(myZip.filenames,myZip.files)
            for myCsvFileName, myCsvFile in csvInZipFile:
                with open(os.path.join(dst,myfile_name),'wb') as outfile:
                    outfile.write(myCsvFile.read())
```
