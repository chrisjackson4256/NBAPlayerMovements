```python
import math
import pandas as pd
```

```python
json_data = pd.read_json("01.05.2016.GSW.at.LAL.json")
```

```python
def get_roster(team):
    team_info = json_data.events[10][team]
    
    team_abbrev = team_info['abbreviation']
    team_name = team_info['name']
    team_id = team_info['teamid']

    team_players = pd.DataFrame(team_info['players'])
    team_players.columns = ['first_name', 'jersey', 'last_name', 'player_id', 'position']
    team_players['team_name'] = team_name
    team_players['team_abbrev'] = team_abbrev
    team_players['team_id'] = team_id

    team_players = team_players[['team_abbrev', 'player_id', 'jersey', 'last_name', 'first_name', 'position']]

    return team_players

home_players = get_roster("home")
visitor_players = get_roster("visitor")

print(home_players)
print()
print(visitor_players)
```

```python
moments = json_data.events[13]['moments']

event_df = pd.DataFrame()
for i in range(len(moments)):

    game_clock = moments[i][2]
    game_minutes = str(math.floor(game_clock / 60.))
    game_seconds = str(round(game_clock - int(game_minutes) * 60));
    if len(game_seconds) < 2:
        game_seconds = "0" + game_seconds
    shot_clock = str(math.floor(moments[i][3]))

    print("Game clock --", game_minutes + ":" + game_seconds)
    print("Shot clock --", shot_clock)

    locations = pd.DataFrame(moments[i][5])
    locations.columns = ['team_id', 'player_id', 'x', 'y', 'z']

    locations['game_minutes'] = game_minutes
    locations['game_seconds'] = game_seconds
    locations['shot_seconds'] = shot_clock

    locations['shot_seconds'].fillna(value='24', inplace=True)

    locations = locations[['game_minutes', 'game_seconds', 'shot_seconds', 'player_id', 'x', 'y', 'z']]
    
    event_df = event_df.append(locations)

event_df
```
